**Ejercicio 1 - Carga de un archivo CSV desde Drive**


```python
import pandas as pd
```

```python
datos = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vT0qy7jVMPTsZa2vAnZV1g_LSrD3Ce8_93qRYvArmP4X0L0mFqtCBynbdos7wYu52Q20xO7wqL28CUB/pub?gid=214651214&single=true&output=csv")
```

```python
datos
```

**Funcion suma a la columna ingresos**

```python
suma_ingresos  = datos['ingresos'].sum()
print("La suma de la columna ingresos es:", suma_ingresos)
```

**Funcion para el promedio de la columna edad**

```python
promedio = datos["edad"].mean()
print("El promedio de edad de los trabajadores es:", int(promedio))
```

**Filtrado de personas por lugar de trabajo**

```python
personas_arequipa = datos[datos.eq("Arequipa").any(axis=1)]
columnas_seleccionadas = personas_arequipa[['nombre', 'ciudad']]
print(columnas_seleccionadas)
```

**Obtenemos el promedio de ingresos por cuidad**

```python
promedio_ingresos_por_ciudad = datos.groupby('ciudad')['ingresos'].mean().astype(int)
print(promedio_ingresos_por_ciudad)
```

**Filtrado de la persona mas joven y mas mayor**

```python
persona_menor_edad = datos.loc[datos['edad'] == datos['edad'].min()].values[0]
persona_mayor_edad = datos.loc[datos['edad'] == datos['edad'].max()].values[0]

print(f"La persona más joven es: {persona_menor_edad}")
print(f"La persona más mayor es: {persona_mayor_edad}")
```

**Numero de trabajadores por cuidad**

```python
personas_por_ciudad = datos.groupby('ciudad')['nombre'].count()
total_trabajadores = personas_por_ciudad.sum()

print(personas_por_ciudad)
print(f"Total de trabajadores: {total_trabajadores}")
```

**Ordenamiento en forma desendiente**

```python
datos.sort_values(by=['edad'], ascending=False)
```

**Promedio de  ingresos de personas mayores a 30 años**

```python
promedio_ingresos_mayores_30 = datos.loc[datos['edad'] > 30, 'ingresos'].mean().astype(int)
print(f"El promedio de ingresos de las personas mayores de 30 años es: {promedio_ingresos_mayores_30}")
```

**Ingresos totales por cuidades**

```python
ingresos_totales_por_ciudad = datos.groupby('ciudad')['ingresos'].sum()
print(ingresos_totales_por_ciudad)
```

**Personas con el ingreso mas alto**

```python
persona_mayor_ingreso = datos.loc[datos['ingresos'] == datos['ingresos'].max()].values[0]
print(f"La persona con el ingreso más alto es: {persona_mayor_ingreso}")
```

**Histograma de edades**

```python
import matplotlib.pyplot as plt
edades = datos["edad"]

plt.hist(edades)
plt.xlabel("Edad")
plt.ylabel("Número de trabajadores")
plt.title("Histograma de edades")
plt.show()
```

**Agregando una columna de impuestos que represente el 10% de ingresos**

```python
datos['impuestos'] = datos['ingresos'] * 0.1
datos.head()
```
